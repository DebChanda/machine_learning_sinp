```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def dist(loc, size = 1000):
    r = np.random.normal(loc=loc, scale=(loc + np.random.rand(1))/4 , size=size)
    theta = np.random.rand(size) * np.pi * 2
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    plt.scatter(x,y)    
    return x,y
```

```python
a = dist(0)
b = dist(3)
```
